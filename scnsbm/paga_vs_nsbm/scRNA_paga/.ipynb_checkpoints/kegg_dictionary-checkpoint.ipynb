```python
def to_df(result):
    return pd.read_table(io.StringIO(result), header=None)
```

# lista con tutti i pathway

```python
result = REST.kegg_list("pathway",'hsa').read()
```

```python
path=to_df(result)
path.shape
```

```python
paths=[i[5:] for i in path.iloc[:,0]]
```

```python
paths[0:5]
```

# dizionario con geni per ogni pathway

```python
pathway = REST.kegg_get("hsa00010").read()
print(pathway)
```

```python
path_words=pathway.split()
```

```python
path_words
```

```python
general_dict=dict()
for path in paths:
    pathway = REST.kegg_get(path).read()
    path_words=pathway.split()
# dalle informazioni del pathway, seleziono solo quelle relative ai geni
    path_genes=list()
    for line in range(len(path_words)):
        if path_words[line]=='GENE':
            start_genes=line+1
        elif path_words[line]=='COMPOUND':
            end_genes=line
    genes=path_words[start_genes:end_genes]
# cerco tutti i termini che sono solo in uppercase e numeri, ma iniziano con una lettera (quelli sono i gene names)
    for line in range(len(genes)):                        
        x=re.findall('[A-Z][A-Z0-9]{2,}', genes[line])
        if len(x)>0:
            if not genes[line].startswith('['):
                path_genes.append(genes[line][0:-1])
# con la chiave del pathway aggiungo al dizionario i geni del pathway
    general_dict[str(path)]=general_dict.get(str(path),path_genes)
```
