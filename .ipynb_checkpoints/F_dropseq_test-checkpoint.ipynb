```python
import scanpy as sc
import pandas as pd
import pickle
import numpy as np
import graph_tool.all as gt
%matplotlib inline
%pylab
```

```python
import anndata
adata=sc.read('adata_dropseq/adata_sc_dropseq.h5ad')
sc.tl.rank_genes_groups(adata, 'leiden', method='t-test')
df_significativo= pd.DataFrame(adata.uns['rank_genes_groups']['names']).head(20) #### df con i 20 geni significativi per gruppo
cell_annotation= pd.read_table('../sc_mixology/data/csv/sc_dropseq.metadata.csv.gz', compression="gzip", sep=",", index_col=0)
adata.obs.loc[:, "cell_line"] = cell_annotation.loc[adata.obs_names, "cell_line"]
```

```python
g = gt.Graph(directed=False)
adj=adata.uns['neighbors']['connectivities']
edge_weights = g.new_edge_property('double')
g.edge_properties['weight'] = edge_weights
nnz = np.nonzero(np.triu(adj.todense(),1))
nedges = len(nnz[0])
g.add_edge_list(np.hstack([np.transpose(nnz),np.reshape(adj[nnz],(nedges,1))]),eprops=[edge_weights])
```

```python
pickle.dump(g,open("adata_dropseq/dropseq_graph_g.pkl",'wb'),2)
```

```python
g = pickle.load( open( "adata_dropseq/dropseq_graph_g.pkl", "rb" ))
```

# A) scelgo il modello tramite entropia da minimize_nested_blockModel
### A1) scelgo modello con entropia minima

```python
state10=[]
for i in range(10):
    i=gt.minimize_nested_blockmodel_dl(g, deg_corr=False)
    state10.append(i)
```

```python
for i in range(10):
    adata.obs.loc[:, 'NSBM_test'] = pd.Categorical(list(state10[i].project_level(1).get_blocks()))
    sc.pl.umap(adata, color=["NSBM_test", 'cell_line'], legend_loc='on data')
    print(pd.crosstab(adata.obs.NSBM_test, adata.obs.cell_line))
```

```python
ent=[]
for i in range(len(state10)):
    x=state10[i].entropy()
    ent.append(x)
```

```python
stateX=None
entX=None
for i in range(len(state10)):
    if stateX==None:
        stateX=state10[i]
        entX=ent[i]
    elif ent[i]<entX:
        entX=ent[i]
        stateX=state10[i]
print(stateX,entX)
```

### A2) trovo grafo con entropia max

```python
stateM=None
entM=None
for i in range(len(state10)):
    if stateM==None:
        stateM=state10[i]
        entM=ent[i]
    elif ent[i]>entM:
        entM=ent[i]
        stateM=state10[i]
print(stateM,entM)
```

```python
pickle.dump(stateX,open("adata_dropseq/F_dropseq_test/state_a_min.pkl",'wb'),2)
```

# C) Utilizzo MCMC_equilibrate per utilizzare la callback
### C1) state_min_sweep

```python
bs = state_min_sweep.get_bs()                   
bs += [np.zeros(1)] * (10 - len(bs))

state = state_min_sweep.copy(bs=bs, sampling=True)
gt.mcmc_equilibrate(state, wait=1000, mcmc_args=dict(niter=10))
h = [np.zeros(g.num_vertices() + 1) for s in state.get_levels()]
pv = [None] * len(state.get_levels())


def collect_info(s):
    for l,sl in enumerate(s.get_levels()):
        B=sl.get_nonempty_B()
        h[l][B] +=1
        global pv
        pv=[sl.collect_vertex_marginals(pv[l]) for l, sl in enumerate(s.get_levels())]

def collect_num_groups(s):
    for l, sl in enumerate(s.get_levels()):
       B = sl.get_nonempty_B()
       h[l][B] += 1
        

def collect_marginals(s):
   global pv
   pv = [sl.collect_vertex_marginals(pv[l]) for l, sl in enumerate(s.get_levels())]
```

```python
bs = state_min_sweep.get_bs()                   
bs += [np.zeros(1)] * (10 - len(bs))

state = state_min_sweep.copy(bs=bs, sampling=True)
gt.mcmc_equilibrate(state_min_sweep, force_niter=10000, mcmc_args=dict(niter=10),callback=collect_info)
```

```python
state
```

```python
adata.obs.loc[:, 'NSBM_test'] = pd.Categorical(list(state.project_level(3).get_blocks()))
sc.pl.umap(adata, color=["NSBM_test", 'cell_line'], legend_loc='on data')
print(pd.crosstab(adata.obs.NSBM_test, adata.obs.cell_line))
```

```python
for l in range (0,len(bs)):
    idxs= np.where(h[l] > 0)[0]
    v = h[l][idxs]
    figure(l)
    bar(idxs, v)
```

```python
pickle.dump(state,open("adata_dropseq/F_dropseq_test/state_c_min_sweep_e.pkl",'wb'),2)
pickle.dump(h,open("adata_dropseq/F_dropseq_test/h_c_min_sweep_e.pkl",'wb'),2)
pickle.dump(pv,open("adata_dropseq/F_dropseq_test/pv_c_min_sweep_e.pkl",'wb'),2)
```
